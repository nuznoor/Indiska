```python
# importing all libraries and the file being used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.linear_model import LinearRegression

filename = r"sales.csv"
all_data = pd.read_csv(filename)
```

```python
def loss_nan(data):
    """Drop "NaN" values from data to not cause errors in calculations"""
    data = data.dropna(how="all")
    return data


all_data = loss_nan(all_data)
```

```python
# Converting date values to datetime format
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
all_data["Ship Date"] = pd.to_datetime(all_data["Ship Date"])
```

```python
# Adding new columns
all_data["Month"] = pd.to_datetime(all_data["Order Date"]).dt.month
all_data["Hour"] = pd.to_datetime(all_data["Order Date"]).dt.hour
```

```python
#Which month sells the most amount of goods
months = range(1, 13)
print(months)

plt.bar(months, all_data.groupby(["Month"]).sum()["Order Quantity"])
plt.xticks(months)
plt.ylabel("Quantity")
plt.xlabel("Month number")
plt.show()
```

```python
# Which Region is most most profitable
region = [city for city, df in all_data.groupby("Region")]
plt.bar(region, all_data.groupby(["Region"]).sum()["Profit"])
plt.xticks(region, rotation="vertical", size=8)
plt.ylabel("Profit in USD ($)")
plt.xlabel("City Name")
plt.show()
```
